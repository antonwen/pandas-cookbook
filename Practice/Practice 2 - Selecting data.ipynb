```python
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 100)
pd.set_option('display.width', 5)
```

```python
complaints = pd.read_csv('./../data/311-service-requests.csv',
                        date_parser='Created Date')
```

```python
complaints[['Unique Key','Created Date']]
```

```python
complaints[['Unique Key', 'Created Date']][90:100]
```

```python
complaint_types = complaints['Complaint Type']
```

```python
complaint_count = complaint_types.value_counts()
```

```python
#print(plt.style.available)
plt.figure(figsize=(10,5))
complaint_count[:10].plot(kind='bar')

plt.show()
```
