```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (15,5)
```

```python
bikes = pd.read_csv('./../data/bikes.csv', sep=';', encoding='latin1', parse_dates=['Date'], index_col=['Date'], dayfirst=True)
```

```python
bikes['Berri 1'].plot()
plt.show()
```

```python
berri_bikes = bikes[['Berri 1']].copy()
```

```python
berri_bikes['weekend'] = berri_bikes.index.weekday

```

```python
berri_group = berri_bikes.groupby('weekend')
berri_group.aggregate('sum')
```
