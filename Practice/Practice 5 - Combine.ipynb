```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (15,5)
```

```python
wheather = pd.read_csv('./../data/weather_2012.csv', 
                       parse_dates=['Date/Time'], 
                       index_col='Date/Time')
```

```python
wheather['Temp (C)'].plot(figsize=(10,2))
```

```python
url_template = "http://climate.weather.gc.ca/climateData/bulkdata_e.html?format=csv&stationID=5415&Year={year}&Month={month}&timeframe=1&submit=Download+Data"
url = url_template.format(year=2013, month=3)
pd.read_csv(url, skiprows=15, index_col='Date/Time', parse_dates=True, header=True)
```

```python
temperatures = wheather[['Temp (C)']].copy()
temperatures['Hour'] = temperatures.index.hour
```

```python
temperatures.groupby('Hour').aggregate(np.mean).plot()
```
